```python
import matplotlib.pyplot as plt

def statis(data):
    if data.dtypes == 'O':
        data.value_counts().plot(kind='bar')
        plt.show()
        data.value_counts().plot(kind='pie', autopct="%1.2f%%")
        plt.show()
    else:
        print("Maximum Value:", data.max())
        print("Minimum Value:", data.min())
        print("Mean Value:", data.mean())
        print("Median Value:", data.median())
        print("Standard Deviation:", data.std())

        data.plot(kind='box')
        plt.show()
        data.plot(kind='kde')
        plt.show()
```

```python
for column_name in df.columns:
    print(f"***{'*'*60}***{column_name} Column***{'*'*60}***")
    statis(df[column_name])
```

**Data Preprocessing**

```python
df.drop(columns=['Like','Age','VisitFrequency','Gender'], inplace=True)
```

```python
df.head(3)
```

```python
X = df.replace({'Yes': 1, 'No': 0})
```

```python
X.head()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=11)
```

```python
pca_result = pca.fit_transform(X)
```

```python
pca_result
```

```python
pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10','PC11'])
```

```python
pca_df.head()
```

```python
from sklearn.cluster import KMeans
```

```python
X = pca_df.iloc[:,:].values
X
```

```python
wcss = [] # empty list
for index in range(1,10): # range 1 to 10 implies that between this range any no. of clusters can be formed
    kmeans = KMeans(n_clusters = index, init = "k-means++", random_state = 2)
    kmeans.fit(X)

    wcss.append(kmeans.inertia_)
plt.plot(range(1,10), wcss)
plt.title("K-Means Clustering")
plt.xlabel("Number of Cluster")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init="k-means++",random_state=42)
```

```python
cluster_values = kmeans.fit_predict(X)
```

```python
# Scatter plot for K-means Clustering

plt.figure(figsize=(8,6))

plt.scatter(X[cluster_values == 0, 2], X[cluster_values == 0, 3], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[cluster_values == 1, 2], X[cluster_values == 1, 3], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[cluster_values == 2, 2], X[cluster_values == 2, 3], s = 100, c = 'green', label = 'Cluster 3')
# plt.scatter(X[cluster_values == 3, 2], X[cluster_values == 3, 3], s = 100, c = 'cyan', label = 'Cluster 4')
# plt.scatter(X[cluster_values == 4, 2], X[cluster_values == 4, 3], s = 100, c = 'magenta', label = 'Cluster 5')

# Centroids for Clusters
plt.scatter(kmeans.cluster_centers_[:, 2], kmeans.cluster_centers_[:, 3], s = 300, c = 'yellow', label = 'Centroids')

# Graph Labels
plt.title('K-means Clustering')
plt.legend()
plt.show()
```
