```python
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.cluster import KMeans
```

```python
# Categorical columns
categorical_columns = ['bodyType', 'fuel']

# Subset the DataFrame with the categorical columns
df_subset = df[categorical_columns]

# Create an instance of the OneHotEncoder
encoder = OneHotEncoder(sparse=False)

# Fit and transform the subset DataFrame
one_hot_encoded = encoder.fit_transform(df_subset)

# Get the unique values in each categorical column
unique_values = [df[column].unique() for column in categorical_columns]

# Create new column names for the one-hot encoded columns
new_columns = []
for i, column in enumerate(categorical_columns):
    for value in unique_values[i]:
        new_columns.append(f"{column}_{value}")

# Create a DataFrame from the one-hot encoded array
one_hot_df = pd.DataFrame(one_hot_encoded, columns=new_columns)

# Concatenate the original DataFrame with the one-hot encoded DataFrame
X = pd.concat([df, one_hot_df], axis=1)
```

```python
wcss = [] # empty list
for index in range(1,11): # range 1 to 10 implies that between this range any no. of clusters can be formed
    kmeans = KMeans(n_clusters = index, init = "k-means++", random_state = 2)
    kmeans.fit(X)
    
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11), wcss)
plt.title("K-Means Clustering")
plt.xlabel("Number of Cluster")
plt.ylabel("WCSS")
plt.show()
```

# **Applying PCA**

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
```

```python
# Define a range of components to loop over
component_range = range(1, X.shape[1] + 1)

best_n_components = None
best_explained_variance_ratio = 0.0

# Loop over different numbers of components
for n_components in component_range:
    # Create an instance of PCA with the current number of components
    pca = PCA(n_components=n_components)

    # Fit the data to the PCA model
    pca.fit(X)

    # Access the explained variance ratio
    explained_variance_ratio = pca.explained_variance_ratio_

    # Check if the current explained variance ratio is better than the best so far
    if explained_variance_ratio[0] > best_explained_variance_ratio:
        best_explained_variance_ratio = explained_variance_ratio[0]
        best_n_components = n_components

# Print the best number of components and the corresponding explained variance ratio
print("Best Number of Components:", best_n_components)
print("Best Explained Variance Ratio:", best_explained_variance_ratio)
```

```python
n_components = 11
pca = PCA(n_components=n_components)
```

```python
# Fit the data to the PCA model

pca.fit(X)
```

```python
# Transform the data to the lower-dimensional representation

X_transformed = pca.transform(X)
```

```python
# Access the principal components

components = pca.components_
```

```python
# Access the explained variance ratio

explained_variance_ratio = pca.explained_variance_ratio_
```

```python
# Print the transformed data, components, and explained variance ratio

print("Transformed Data:")
print(X_transformed)
print("\nPrincipal Components:")
print(components)
print("\nExplained Variance Ratio:")
print(explained_variance_ratio)
```

```python
wcss = [] # empty list
for index in range(1,11): # range 1 to 10 implies that between this range any no. of clusters can be formed
    kmeans = KMeans(n_clusters = index, init = "k-means++", random_state = 2)
    kmeans.fit(X_transformed)
    
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11), wcss)
plt.title("After Applying PCA")
plt.xlabel("Number of Cluster")
plt.ylabel("WCSS")
plt.show()
```
